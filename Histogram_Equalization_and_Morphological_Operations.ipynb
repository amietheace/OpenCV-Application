```python
# Morphological can be best understood on Black and white image
# Thats why we have created one in the above code
# Erision is a type of Morphology
""" like erosion does in geography in image processing also
it will erode pixels"""

import cv2
import matplotlib.pyplot as plt
import numpy as np


def main():
    
    path = "D:\\CV_and_NLP\\opencv\\dataset\\"

    imgpath1 =  path + "cameraman.tif"
    


    img = cv2.imread(imgpath1, 0)
    
#    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    th = 0
    max_val = 255
    

    ret, binary_inv = cv2.threshold(img, th, max_val, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    
    k = np.ones((5, 5), np.uint8)
    erosion = cv2.erode(binary_inv, k, iterations=3)
    
    output = [img, binary_inv, erosion]
    
    titles = ['Original', 'Binary Inv', 'Erosion']
    
    for i in range(3):
        plt.subplot(1, 3, i+1)
        plt.imshow(output[i], cmap='gray')
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])

    plt.show()  

if __name__ == "__main__":
    main()
```

```python
"""Dialtion is also a type of morphology, it is reverse of 
erosion,"""

import cv2
import matplotlib.pyplot as plt
import numpy as np


def main():
    
    path = "D:\\CV_and_NLP\\opencv\\dataset\\"

    imgpath1 =  path + "cameraman.tif"
    


    img = cv2.imread(imgpath1, 0)
    
#    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    th = 0
    max_val = 255
    

    ret, binary_inv = cv2.threshold(img, th, max_val, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    
    k = np.ones((5, 5), np.uint8)
    erosion = cv2.erode(binary_inv, k, iterations=3)
    
    dilation = cv2.dilate(binary_inv, k, iterations=3)
    
    output = [img, binary_inv, erosion, dilation]
    
    titles = ['Original', 'Binary Inv', 'Erosion', 'Dilation']
    
    for i in range(4):
        plt.subplot(2, 2, i+1)
        plt.imshow(output[i], cmap='gray')
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])

    plt.show()  

if __name__ == "__main__":
    main()
```

```python
"""Gradient is also a morphological operation, it the difference between 
Erosion and Dilation (with same no of iterations)"""

import cv2
import matplotlib.pyplot as plt
import numpy as np


def main():
    
    path = "D:\\CV_and_NLP\\opencv\\dataset\\"

    imgpath1 =  path + "cameraman.tif"
    


    img = cv2.imread(imgpath1, 0)
    
#    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    th = 0
    max_val = 255
    

    ret, binary_inv = cv2.threshold(img, th, max_val, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    
    k = np.ones((5, 5), np.uint8)
    erosion = cv2.erode(binary_inv, k, iterations=3)
    
    dilation = cv2.dilate(binary_inv, k, iterations=3)
    
    gradient = cv2.morphologyEx(binary_inv, cv2.MORPH_GRADIENT, k)
    
    output = [img, binary_inv, erosion, dilation, gradient]
    
    titles = ['Original', 'Binary Inv', 'Erosion', 'Dialation', 'Gradient']
    
    for i in range(5):
        plt.subplot(3, 2, i+1)
        plt.imshow(output[i], cmap='gray')
        plt.title(titles[i])
        plt.xticks([])
        plt.yticks([])

    plt.show()  

if __name__ == "__main__":
    main()
```
